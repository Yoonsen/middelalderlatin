## Funksjoner for å se på data

```python
from unidecode import unidecode

def find_lemma_occurrences(df = lemmas, lemma = "aqua"):
    return df[df.no_accent == lemma]

def find_form_occurences(df = lemmas, form = "aqua"):
    return df[df.FORM == form]

def find_forms_of_lemma(df = lemmas, lemma = "aqua"):
    return set(list(df.FORM[df.no_accent == lemma]))
```

## I en pandastruktur

```python
lemmas['no_accent'] = lemmas['LEMMA']
```

```python
series = map(unidecode,lemmas['no_accent'].fillna(''))
```

```python
whole_series = [x for x in series]
```

```python
lemmas['no_accent'] = whole_series
```
