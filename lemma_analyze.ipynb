```python
def find_lemma(df = lemmas, lemma = "aqua"):
    return df[df.LEMMA == lemma]

def find_forms_of_lemma(df = lemmas, lemma = "aqua"):
    return set(list(df.FORM[df.LEMMA == lemma]))
```
