# Konkordanser fra DN

#### Aktiver cellen under for å starte (trykk ctrl-enter)

Kommentar til selv

```python


def dn_conc(word, param={
             'size':10, 
             'offset':1, 
             'bind':0, 
             'window':5}):
    import re
    from IPython.display import HTML
    import requests
    
    param['query'] = word
    r = requests.get('https://api.nb.no/korpus/latin', params=param)
    res = []
    if r.status_code == 200:
        res = r.text
    if res == []:
        res = "<p>Ingen treff</p>"
    return HTML(res)




def dn_freq(word):
    import re
    from IPython.display import HTML
    import requests
    
    r = requests.get('https://api.nb.no/korpus/freq?query={q}'.format(
        q=word))
    res = []
    if r.status_code == 200:
        res = r.text
    if res == []:
        res = "<p>Ingen treff</p>"
    return HTML(res)

```

#### Konkordanser lages med kommandoen __dn_conc__(_søkeuttrykk_). Lag nye celler for å sammenligne konkordanser, eller kjør forskjellige konkordanser i samme vindu.

#### Søkeuttrykket kan være et ord som _omni_:

```python
dn_conc('baroni.*')
```

#### Søk med regulære uttrykk ved å sette in kombinasjoner av punktum '.' og stjerne '*' 

```python
dn_conc('abd.ct.')
```

```python
dn_freq('abbas')
```

#### Begrens antallet treff med parameteret _size_ , og skip over treff med _offset_

```python
dn_conc('u.*', param = { 'size':2, 'offset':0})
```

### Begrens til et bestemt bind med 'bind':N

```python
dn_conc('aq.*', param = { 'bind':17, 'size':10})
```
